```python
import numpy as np
import os
import pandas as pd
import xarray as xr

import cartopy, cartopy.crs as ccrs
import geopandas
import matplotlib, matplotlib.pyplot as plt
import rasterio
import rioxarray
from shapely.geometry import mapping
```

```python
def load_EIA_data(data_dirname: str):

    ''' 
    Loads EIA energy data into an xArray Dataset and averages it based on the given string.
    '''

    # Load ERA5 data
    EIA_filename = 'dataset-EIA.field_name-net_energy_load.start_date-19900101.end_date-20090101.nc'
    EIA_dataset = xr.open_dataset(os.path.join(data_dirname, EIA_filename))

    date_range = pd.date_range('1990-01-01', '2010-01-01', freq='M')
    EIA_dataset['year'] = pd.date_range('1990-01-01', '2010-01-01', freq='Y')
    EIA_dataset = EIA_dataset.interp(year=date_range, method='quadratic').rename({'year': 'time'})
    
    return EIA_dataset
```

```python
def load_census_data(data_dirname: str):
    filename = 'census_nhgis.csv'
    TEMP = pd.read_csv(os.path.join(data_dirname, filename))
    
    container = []
    for nerc_region, nerc_population in TEMP.groupby('nerc'):
        nerc_population = nerc_population.reset_index(drop=True).drop(columns=['state', 'nerc']).T
        nerc_population = pd.DataFrame(nerc_population.sum(axis=1), columns=['population'])
        nerc_population['region'] = nerc_region
        nerc_population = nerc_population.reset_index(names=['year'])
        nerc_population['year'] = nerc_population['year'].astype(int)
        container.append(nerc_population)
    nerc_population = pd.concat(container)

    TMP = nerc_population.groupby('year').sum().reset_index()
    TMP['region'] = 'ALL_US'
    nerc_population = pd.concat([nerc_population, TMP])
    
    nerc_population = nerc_population.set_index(['region', 'year']).to_xarray()
    nerc_population = nerc_population.interp(year=np.arange(1960, 2020, 1))
    
    date_range = pd.date_range('1960-01-01', '2020-01-01', freq='M')
    nerc_population['year'] = pd.date_range('1960-01-01', '2020-01-01', freq='Y')
    nerc_population = nerc_population.interp(year=date_range, method='quadratic').rename({'year': 'time'})
    
    return nerc_population
```

```python
def load_ERA5_data(data_dirname: str,
                   data_averaging: str|None='Y'):

    ''' 
    Loads ERA5 climate data into an xArray Dataset and averages it based on the given string.
    '''

    # Load ERA5 data
    ERA5_filename = 'dataset-ERA5.field_name-t2m.start_date-19900101.end_date-20141231.nc'
    ERA5_dataset = xr.open_dataset(os.path.join(data_dirname, ERA5_filename))
    # Rename time axis and temperature variables
    ERA5_dataset = ERA5_dataset.rename({'valid_time': 'time', 't2m': 'T'})

    # Perform data averaging
    if data_averaging is not None:
        ERA5_dataset = ERA5_dataset.resample(time='1M').mean('time').resample(time=data_averaging).mean('time')
    
    return ERA5_dataset
```

```python
def load_CMIP6_data(data_dirname: str|None=None):

    if data_dirname == None:
        data_dirname = '/Users/gabriel/Documents/princeton/courses/spring_2025/cos_513/assignments/project/data'
    
    container = []
    temperature_varnames = []
    identifier = 'cmip6_monthly'
    
    for scenario in ['historical', 'ssp126', 'ssp585']:
        pathnames = [os.path.join(data_dirname, f) for f in os.listdir(data_dirname) if
                     f.endswith('.nc') and 
                     identifier in f and
                     scenario in f]
        
        for pathname in pathnames:
            temperature_varname = f"T.{scenario}.{pathname.split('/')[-1].split('.')[-3]}"
            TMP = xr.open_dataset(pathname).rename({'tas': temperature_varname}).convert_calendar('standard', dim='time', align_on='year', use_cftime=None)
            container.append(TMP)
            del TMP
            temperature_varnames.append(temperature_varname)
        
    dataset = xr.merge(container, compat='override')

    for temperature_varname in temperature_varnames:
        dataset[temperature_varname] = dataset[temperature_varname].interpolate_na('time')

    dataset = dataset.drop(['time_bnds', 'lat_bnds', 'lon_bnds', 'time_bounds']).resample(time='1M').mean('time')

    return dataset
```

```python
def subdivide_ERA5_data(ERA5_dataset: xr.Dataset,
                        data_dirname: str):

    '''
    Function to mask ERA5 data into NERC regions.
    
    Shapefile asset credit: 
    - Organization name: U.S. Department of Homeland Security, 
    - URL: https://hifld-geoplatform.hub.arcgis.com/datasets/geoplatform::nerc-regions/about 
    '''
    
    # Load North American Electric Reliability Corporation (NERC) shapefiles
    dirname_shapefiles = f'{data_dirname}/nerc_region_shapefiles'
    pathname_shapefiles = os.path.join(dirname_shapefiles, 'NERC_Regions_Subregions.shp')
    nerc_shapefile = geopandas.read_file(pathname_shapefiles, crs="epsg:4326")
    
    # Get region codes corresponding to subdivision acronyms
    # These acronyms are used as top-level identifiers in the US EIA energy load data
    nerc_shapefile['REGION_CODE'] = nerc_shapefile['NAME'].str.extract('\((.*?)\)')
    nerc_shapefile.loc[nerc_shapefile.REGION_CODE == 'TRE', 'REGION_CODE'] = 'ERCOT'
    
    # Re-align ERA5 data with NERC geospatial mapping
    ERA5_dataset.rio.set_spatial_dims(x_dim='longitude', y_dim='latitude', inplace=True)
    ERA5_dataset.rio.write_crs('epsg:4326', inplace=True)
    
    # Iterate through all NERC region codes to mask regional ERA5 data
    container, masks = {}, {}
    for nerc_region_name, nerc_region in nerc_shapefile.groupby('REGION_CODE'):
        # Clip data to iterand NERC region
        masked_ERA5_dataset = ERA5_dataset.rio.clip(nerc_region.geometry.apply(mapping), nerc_region.crs, drop=False)
        masked_ERA5_dataset = masked_ERA5_dataset.assign_coords(region=nerc_region_name).expand_dims('region')
        # Rename data variables in ERA5 dataset to appropriate the data to the iterand NERC region
        # masked_ERA5_dataset = masked_ERA5_dataset.rename({v: f'{v}-{nerc_region_name}' for v in masked_ERA5_dataset.data_vars})
        container[nerc_region_name] = masked_ERA5_dataset.mean(['latitude', 'longitude'])
        masks[nerc_region_name] = xr.where(~np.isnan(masked_ERA5_dataset), True, False)
    
    # Get ERA5 data for all NERC regions
    nerc_region_name = 'ALL_US'
    masked_ERA5_dataset = ERA5_dataset.rio.clip(nerc_shapefile.geometry.apply(mapping), nerc_shapefile.crs, drop=False)
    masked_ERA5_dataset = masked_ERA5_dataset.assign_coords(region=nerc_region_name).expand_dims('region')
    container[nerc_region_name] = masked_ERA5_dataset.mean(['latitude', 'longitude'])
    masks[nerc_region_name] = xr.where(~np.isnan(masked_ERA5_dataset), True, False)

    # Merge all the datasets
    dataset = xr.merge(container.values())

    return dataset, masks
```

```python
def interpolate_CMIP6_data(CMIP_dataset: xr.Dataset,
                           ERA5_basis_vector_x: np.array,
                           ERA5_basis_vector_y: np.array):

    # Adjust convention from (-180, 180) to (0, 360)
    ERA5_basis_vector_x = ERA5_basis_vector_x + 360
    ERA5_basis_vector_y = ERA5_basis_vector_y
    # Iterate over all CMIP members and assign to a dictionary
    CMIP_dataset_interpolated = {}
    for CMIP_member in CMIP_dataset.data_vars:
        CMIP_dataset_interpolated[CMIP_member] = CMIP_dataset[CMIP_member].interpolate_na(dim='lat').interpolate_na(dim='lon').interp(lon=ERA5_basis_vector_x).interp(lat=ERA5_basis_vector_y)
    
    return xr.merge(CMIP_dataset_interpolated.values())
```

```python
def subdivide_CMIP6_data(CMIP6_dataset: xr.Dataset,
                         data_dirname: str):

    '''
    Function to mask ERA5 data into NERC regions.
    
    Shapefile asset credit: 
    - Organization name: U.S. Department of Homeland Security, 
    - URL: https://hifld-geoplatform.hub.arcgis.com/datasets/geoplatform::nerc-regions/about 
    '''
    
    # Load North American Electric Reliability Corporation (NERC) shapefiles
    dirname_shapefiles = f'{data_dirname}/nerc_region_shapefiles'
    pathname_shapefiles = os.path.join(dirname_shapefiles, 'NERC_Regions_Subregions.shp')
    nerc_shapefile = geopandas.read_file(pathname_shapefiles, crs="epsg:4326")
    
    # Get region codes corresponding to subdivision acronyms
    # These acronyms are used as top-level identifiers in the US EIA energy load data
    nerc_shapefile['REGION_CODE'] = nerc_shapefile['NAME'].str.extract('\((.*?)\)')
    nerc_shapefile.loc[nerc_shapefile.REGION_CODE == 'TRE', 'REGION_CODE'] = 'ERCOT'
    
    # Re-align ERA5 data with NERC geospatial mapping
    CMIP6_dataset.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)
    CMIP6_dataset.rio.write_crs('epsg:4326', inplace=True)
    
    # Iterate through all NERC region codes to mask regional ERA5 data
    container = {}
    masks = {}
    for nerc_region_name, nerc_region in nerc_shapefile.groupby('REGION_CODE'):
        # Clip data to iterand NERC region
        masked_CMIP6_dataset = CMIP6_dataset.rio.clip(nerc_region.geometry.apply(mapping), nerc_region.crs, drop=False)
        masked_CMIP6_dataset = masked_CMIP6_dataset.assign_coords(region=nerc_region_name).expand_dims('region')
        # Rename data variables in CMIP6 dataset to appropriate the data to the iterand NERC region
        # masked_ERA5_dataset = masked_ERA5_dataset.rename({v: f'{v}-{nerc_region_name}' for v in masked_ERA5_dataset.data_vars})
        container[nerc_region_name] = masked_CMIP6_dataset.mean(['lat', 'lon'])
    
    # Get ERA5 data for all NERC regions
    nerc_region_name = 'ALL_US'
    masked_CMIP6_dataset = CMIP6_dataset.rio.clip(nerc_shapefile.geometry.apply(mapping), nerc_shapefile.crs, drop=False)
    masked_CMIP6_dataset = masked_CMIP6_dataset.assign_coords(region=nerc_region_name).expand_dims('region')
    container[nerc_region_name] = masked_CMIP6_dataset.mean(['lat', 'lon'])

    # Merge all the datasets
    dataset = xr.merge(container.values())

    return dataset
```

```python
def load(dirname: str|None = None,
         overwrite: bool=False):
    
    if not dirname:
        dirname = '/Users/gabriel/Documents/princeton/courses/spring_2025/cos_513/assignments/project/data'

    # Define storage path
    storage_filename = 'dataset.population-energy_load-temperature.nc'
    storage_pathname = os.path.join(dirname, storage_filename)

    # Boolean to dictate whether data is already generated and saved
    file_exists = os.path.exists(storage_pathname)

    # Check if the file already exists. If yes and overwriting isn't being requested, load. Else, generate.
    if file_exists and not overwrite:
        print(f'File exists at {storage_pathname}, loading...')
        dataset = xr.open_dataset(storage_pathname)
    else:
        print(f'File does not exist at {storage_pathname}, generating...')
        ''' Load energy data. '''
        EIA_dataset = load_EIA_data(data_dirname=dirname)
        ''' Load census data. '''
        census_dataset = load_census_data(data_dirname=dirname)
        ''' Load ERA5 reanalysis data. '''
        dataset_ERA5 = load_ERA5_data(data_dirname=dirname, data_averaging='M')
        # Subdivide by NERC region and obtain boolean masks
        dataset_ERA5_subdivided, masks = subdivide_ERA5_data(ERA5_dataset=dataset_ERA5,
                                                             data_dirname=dirname)
        ''' Load CMIP6 temperature projection data. '''
        dataset_CMIP6 = load_CMIP6_data()
        boolean_masks = {mask_region: mask['T'].mean(['time', 'region']).rename({'latitude': 'lat', 'longitude': 'lon'}) for mask_region, mask in masks.items()}
        # Interpolate CMIP6 data to the ERA5 grid (any region name can be used to access basis vector values)
        dataset_CMIP6_interpolated = interpolate_CMIP6_data(CMIP_dataset=dataset_CMIP6,
                                                            ERA5_basis_vector_x=boolean_masks['ERCOT'].lon,
                                                            ERA5_basis_vector_y=boolean_masks['ERCOT'].lat)
        dataset_CMIP6_regional = subdivide_CMIP6_data(dataset_CMIP6_interpolated, dirname)
        ''' Unite all datasets. '''
        dataset = xr.merge([dataset_ERA5_subdivided, EIA_dataset, census_dataset, dataset_CMIP6_regional])

        dataset.load().to_netcdf(f'{dirname}/dataset.population-energy_load-temperature.nc')

    return dataset
```

```python
dataset = load(overwrite=False)
```

### Visualization

```python
num_rolling_years = 5
model_names = ['hadgem3_gc31_ll', 'ipsl_cm6a_lr', 'mpi_esm1_2_lr', 'ukesm1_0_ll']
low_projection_name, high_projection_name = 'ssp126', 'ssp585'

region_name = 'NPCC'

fig, ax = plt.subplots(figsize=(5, 3), dpi=144)
for model_name in model_names:
    proj_hist = dataset[f'T.historical.{model_name}'].sel(region=region_name).rolling(time=12 * num_rolling_years, center=False).mean()
    ax.plot(proj_hist.time, proj_hist - 273.15, label=model_name.upper())

plt.gca().set_prop_cycle(None)
for model_name in model_names:
    proj_low = dataset[f'T.{low_projection_name}.{model_name}'].sel(region=region_name).rolling(time=12 * num_rolling_years, center=False).mean()
    proj_high = dataset[f'T.{high_projection_name}.{model_name}'].sel(region=region_name).rolling(time=12 * num_rolling_years, center=False).mean()
    proj_mean = ((proj_low + proj_high)/2)

    timestamps = proj_mean.time - np.timedelta64(86400*365*5, 's')

    ax.plot(timestamps, proj_mean - 273.15, label='_nolegend_')
    ax.fill_between(x=timestamps, y1=proj_low - 273.15, y2=proj_high - 273.15, alpha=0.25)

ax.axvline(proj_low.sel(time='2014-07-01', method='nearest').time.values, c='k', lw=2)

ax.set_title('')
ax.legend(frameon=False, fontsize=7)
ax.set_title(f'CMIP6 ensemble projections, temperature, {region_name}\n{num_rolling_years}-year rolling mean, SSP126 to SSP585', loc='left', ha='left', fontsize=10)
ax.set_ylabel('deg C')
ax.set_xlim([proj_mean.sel(time='1960-01-01', method='nearest').time.min(), proj_mean.sel(time='2095-01-01', method='nearest').time])
```
